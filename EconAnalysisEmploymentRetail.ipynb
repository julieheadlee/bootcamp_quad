# Ecomomic Impact -- Employement, Retail & Medical Sales


In Colorado consumers have the option to access cannabis either for either Medical or Recreational purposes.

As of Dec 1st, 2020 there were a total of 431 medical cannabis retail locations and 603 retail (recreational) locations.  Outlets can choose to sell both medial and recreational cannabis but must have seperate license.



Currently 58.3% of all retail locations are Retail and 41.7% are Medical

Most retail locations are located in the eastern portion of the state in heavily populated areas



The legalization of cannabis has also brought employment to Colorado.  While cannabis jobs are not recorded in Federal databases because cannabis is still illegal at a Federal level employment in the cannabis industry in Colorado requires a license.  As demand for cannabis grows so have the numbers employed in the industry each year from 2017 to 2020
